```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
x_data = [1, 2, 3, 4, 5, 6, 7, 8]
y1 = [87.35, 90.93, 93.85, 96.28, 97, 98, 95, 92] # 11M cell acc
y2 = [10.50, 17, 25, 41, 61, 58, 15, 8.21] # 11M board acc

y3 = [89.36, 92, 93.69, 97, 98, 98, 97, 95] # 21M cell acc
y4 = [13, 18.55, 36.55, 50.74, 68, 68, 30.53, 17.53] # 21M board acc

cmap = cm.plasma
colors = [cmap(0.7), cmap(0.9), cmap(0.2), cmap(0.4)]

#sns.set_context('talk')
sns.set_theme('talk')

#ax = sns.lineplot(x=x_data, y=y1, color=colors[0], linewidth=2, label="slt")
#sns.scatterplot(ax=ax, x=x_data, y=y1, color=colors[0], s=50)

ax = sns.lineplot(x=x_data, y=y2, color=colors[0], linewidth=2, label="Transformer 11M")
sns.scatterplot(ax=ax, x=x_data, y=y2, color=colors[0], s=50)

#sns.lineplot(ax=ax, x=x_data, y=y3, color=colors[2], linewidth=2)
#sns.scatterplot(ax=ax, x=x_data, y=y3, color=colors[2], s=50)

sns.lineplot(ax=ax, x=x_data, y=y4, color=colors[1], linewidth=2, label="Transformer 21M")
sns.scatterplot(ax=ax, x=x_data, y=y4, color=colors[1], s=50)

ax.set_xticks(x_data)
ax.set_ybound(0, 100)

ax.set_title('Board accuracy through layers', fontsize=18)
ax.set_xlabel('layer')
ax.set_ylabel('accuracy')
```
